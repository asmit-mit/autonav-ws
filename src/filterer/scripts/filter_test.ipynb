```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "/home/asmitpaul/Downloads/lanes.jpg"
img = cv2.imread(path)
print("Image Shape:", img.shape)
```

```python
plt.imshow(img)
```

```python
hsv_image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

upper_hsv = np.array([179, 50, 255])
lower_hsv = np.array([0, 0, 220])

hsv_mask = cv2.inRange(hsv_image, lower_hsv, upper_hsv)
```

```python
kernel = np.ones((3, 3), np.uint8)

eroded = cv2.erode(hsv_mask, kernel, iterations=20)
dilated = cv2.dilate(eroded, kernel, iterations=20)
```

```python
plt.imshow(dilated)
```
